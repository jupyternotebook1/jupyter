```python
# Q2

import numpy as np
import matplotlib.pyplot as plt

# Parameters
alpha = 0.3       # Output elasticity of capital
beta = 0.96       # Discount factor
delta = 0.1       # Depreciation rate
T = 50            # Time periods
K0 = 1            # Initial capital
C0 = 0.5          # Initial consumption

# Arrays to store capital and consumption over time
K = np.zeros(T)
C = np.zeros(T)
K[0] = K0
C[0] = C0

# Simulation of the model
for t in range(1, T):
    Y = K[t-1]**alpha  # Output from capital
    C[t] = beta * (Y + (1 - delta) * K[t-1])  # Consumption based on utility maximization
    K[t] = (1 - delta) * K[t-1] + (Y - C[t])  # Capital accumulation

# Plotting results
plt.figure(figsize=(10, 5))
plt.plot(K, label='Capital Stock (K)', color='blue')
plt.plot(C, label='Consumption (C)', color='orange')
plt.title('Capital and Consumption Over Time')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid()
plt.show()
```
